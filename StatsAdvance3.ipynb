Estimation statistics is a branch of statistics that deals with estimating the value of population parameters based on sample data. It provides methods to infer the value of an unknown parameter (such as a population mean or proportion) using sample statistics.

Interval Estimate:
An interval estimate provides a range of values within which the population parameter is expected to lie. This range is often constructed with a certain level of confidence (e.g., 95% confidence interval).

Example:
If a sample mean height of students is 170 cm with a 95% confidence interval of [168, 172] cm, it means we are 95% confident that the true population mean height lies between 168 cm and 172 cm.

```python
import scipy.stats as stats
import numpy as np

def estimate_population_mean(sample, confidence_level):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)
    
    # Calculate the confidence interval
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
    margin_of_error = t_critical * (sample_std / np.sqrt(n))
    
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return sample_mean, (lower_bound, upper_bound)

# Example usage:
sample_data = [50, 60, 55, 70, 65, 75, 80, 85]
confidence_level = 0.95
mean, interval = estimate_population_mean(sample_data, confidence_level)
print(f"Sample Mean: {mean}, Confidence Interval: {interval}")
```

```python
import numpy as np
from scipy import stats

def hypothesis_test(sample1, sample2, alpha=0.05):
    # Calculate the sample means
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    
    # Calculate the standard deviations
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    
    # Calculate the sample sizes
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate the standard error of the difference in means
    se_diff = np.sqrt(std1**2 / n1 + std2**2 / n2)
    
    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / se_diff
    
    # Degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate the critical t-value
    t_critical = stats.t.ppf(1 - alpha / 2, df)
    
    # Calculate the p-value
    p_value = (1 - stats.t.cdf(abs(t_statistic), df)) * 2
    
    return t_statistic, t_critical, p_value

# Example usage
sample1 = [68, 70, 72, 71, 69, 73, 68, 74]
sample2 = [65, 67, 66, 64, 68, 65, 66, 63]

t_stat, t_crit, p_val = hypothesis_test(sample1, sample2)
print(f"T-statistic: {t_stat}, Critical value: {t_crit}, P-value: {p_val}")

# Decision
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    # Calculate the sample means and standard deviations
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)
    std2 = np.std(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate the standard error of the difference in means
    se_diff = np.sqrt(std1**2 / n1 + std2**2 / n2)
    
    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / se_diff
    
    # Degrees of freedom
    df = n1 + n2 - 2
    
    # Calculate the critical t-value
    t_critical = stats.t.ppf(1 - alpha / 2, df)
    
    # Calculate the p-value
    p_value = (1 - stats.t.cdf(abs(t_statistic), df)) * 2
    
    return t_statistic, t_critical, p_value

# Example usage
np.random.seed(0)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=10, size=30)

t_stat, t_crit, p_val = two_sample_t_test(sample1, sample2)
print(f"T-statistic: {t_stat}, Critical value: {t_crit}, P-value: {p_val}")

# Decision
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
